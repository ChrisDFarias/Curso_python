```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp FROM
master
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df=pd.read_sql_query(QUERY,conn)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
```

```python
ax=df['ConvertedComp'].plot(kind='hist')
ax.set_title('Histogram of ConvertedComp')
ax.set_xlabel('ConvertedComp')
ax.set_ylabel('Counts')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age FROM
master
"""
df=pd.read_sql_query(QUERY,conn)

ax=df['Age'].plot(kind='box')
ax.set_title('BoxPlot of Age')

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs FROM
master
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df=pd.read_sql_query(QUERY,conn)

ax=df.plot(kind='scatter',x='Age',y='WorkWeekHrs')
ax.set_title('Scatter plot of Age vs WorkWeekHrs')
ax.set_xlabel('Age')
ax.set_ylabel('WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs,Age FROM
master
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df=pd.read_sql_query(QUERY,conn)
ax=df.plot(kind='scatter',x='WorkWeekHrs',y='CodeRevHrs',s='Age')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT LanguageDesireNextYear,COUNT(*) as Frecuencia FROM
LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Frecuencia DESC
"""
df=pd.read_sql_query(QUERY,conn)
df
```

```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear,COUNT(*) as Frecuencia FROM
DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Frecuencia DESC
LIMIT 5
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df=pd.read_sql_query(QUERY,conn)

plt.pie(x=df['Frecuencia'],labels=df['DatabaseDesireNextYear'],autopct="%.1f%%")
plt.title('Pie chart of top 5 DatabaseDesireNextYear')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs,Age FROM
master
WHERE Age BETWEEN 30 AND 35
"""
df=pd.read_sql_query(QUERY,conn)
df_grouped=df.groupby('Age',as_index=False)['WorkWeekHrs','CodeRevHrs'].median()
df_grouped.plot(kind='bar',x='Age',stacked='True')
```

```python
QUERY = """
SELECT * FROM
LanguageWorkedWith
"""
df=pd.read_sql_query(QUERY,conn)
len(df[df['LanguageWorkedWith']=='SQL'])
```

```python
QUERY = """
SELECT Respondent,DatabaseWorkedWith,COUNT(DatabaseWorkedWith) as Bases FROM
DatabaseWorkedWith
GROUP BY Respondent
"""
df=pd.read_sql_query(QUERY,conn)
len(df[(df['DatabaseWorkedWith']=='MySQL') & (df['Bases']==1) ])
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp,Age FROM
master
WHERE Age BETWEEN 45 AND 60
"""
df=pd.read_sql_query(QUERY,conn)
df_grouped=df.groupby('Age',as_index=False)['ConvertedComp'].median()
df_grouped.plot(kind='line',x='Age')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch,COUNT(MainBranch) as Contador FROM
master
GROUP BY MainBranch
"""
df=pd.read_sql_query(QUERY,conn)
df.plot(kind='barh',x='MainBranch',y='Contador')
```
